# Class 4 — TensorFlow Sample

A minimal end-to-end example: build, train, and evaluate a neural network on MNIST digits.

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

print("TensorFlow version:", tf.__version__)
```

```python
# Load MNIST: 60k training + 10k test images of handwritten digits (28x28 grayscale)
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

# Normalize pixel values to [0, 1] and flatten 28x28 -> 784
x_train = x_train.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0
x_train = x_train.reshape(-1, 28 * 28)
x_test = x_test.reshape(-1, 28 * 28)

print("Training samples:", x_train.shape[0])
print("Test samples:", x_test.shape[0])
```

```python
# Sequential model: input -> hidden layer -> output (10 digit classes)
model = keras.Sequential([
    layers.Input(shape=(28 * 28,)),
    layers.Dense(128, activation="relu"),
    layers.Dense(10, activation="softmax"),
])

model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"],
)

model.summary()
```

```python
# Train for 5 epochs (use a small subset for a quick demo)
history = model.fit(
    x_train[:10000],
    y_train[:10000],
    epochs=5,
    batch_size=128,
    validation_split=0.1,
    verbose=1,
)
```

```python
# Evaluate on held-out test data
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=0)
print(f"Test accuracy: {test_acc:.4f}")
```

```python
# Single prediction example
import numpy as np

sample_idx = 0
prediction = model.predict(x_test[sample_idx : sample_idx + 1], verbose=0)
predicted_digit = int(np.argmax(prediction))

print(f"True label: {y_test[sample_idx]}")
print(f"Predicted:  {predicted_digit}")
print(f"Probabilities: {np.round(prediction[0], 3)}")
```
